### How to do these actions(Retweet, Quote Tweet, Reply)?

- How to RT/Reply/QuoteTweet? I'm not much familiar with twitter.

- There are 3 buttons under a particular twitter tweet.

```
@user_id

....tweet text....

```

- (massage balloon) = **reply**, leave a comment under the tweet
- (2 circular arrows - rt button) = will popup menu to **retweet** or **quote tweet**
- (heart) = **like**

You can also do **quote tweet** by copying a tweet link.

**Quote Tweet** is sharing another's tweet and adding your thought on the top of it.



### How do I know which action(RT/QT/reply) a particular row in this raw data is?

#### RT

- from the `is_retweet` column
- from the `tweet_text` column: it will start with `RT @original_user_id: ...text...`. it shows the original user post.

#### Reply

- from the `in_reply...` column. it will link to the original tweet/user id.

#### QT

- from the `quoted_tweet_tweetid` = not null
- if a user RT the quoted tweet: the `is_retweet` is `True`

### interesting columns

- `is_retweet` and `tweet_text`
     
- the id in `in_reply_to...` and `user_mentions` **are not hashed**, can use it to find more infomation
    - numeric user id
    
- `quoted_tweet..` can tell us if a tweet is tweeted by `quote tweet` action (also RT the **quote tweet** from another person)
